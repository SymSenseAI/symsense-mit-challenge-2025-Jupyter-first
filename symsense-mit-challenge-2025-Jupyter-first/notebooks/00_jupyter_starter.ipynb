```python
##Finishing cleaning after previous crash
clean_path = "data/clean/mit-t08/nhanes/v20251013"
```

```python
#Changing all ints to floats
def standardize_to_float(df: pd.DataFrame) -> pd.DataFrame:
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)
    return df
```

```python
#updating previous cleaned files
for filename in os.listdir(clean_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(clean_path, filename)
        print(f"processing {filename}...")
        df = pd.read_csv(file_path, low_memory = False)
        df = standardize_to_float(df)
        df.to_csv(file_path, index=False)
```
