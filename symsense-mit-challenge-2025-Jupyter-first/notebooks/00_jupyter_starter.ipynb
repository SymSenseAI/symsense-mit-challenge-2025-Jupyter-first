## Configure data path
Set `SYMSENSE_DATA` environment variable to the shared dataset folder or use a local fallback.

```
import os, pandas as pd
DATA_ROOT = os.getenv("SYMSENSE_DATA", "data/raw")
print("DATA_ROOT:", DATA_ROOT)
# Example: try reading a small CSV if present
example = os.path.join(DATA_ROOT, "example.csv")
if os.path.exists(example):
    df = pd.read_csv(example)
    display(df.head())
else:
    print("Place a small CSV at:", example)
```
